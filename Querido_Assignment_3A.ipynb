````python
# cleanrecweekdata.csv
import csv
import re

with open('recweekdata.csv','r') as dirty:
    dirty_reader = csv.reader(dirty, delimiter = ',')
    
    # creating dirty and already clean lists
    clean_firstname = []
    clean_lastname = []
    dirty_birthmonth = []
    clean_birthday = []
    dirty_birthyear = []
    dirty_course = []
    dirty_phonenumber = []
    line_count = 0
    for row in dirty_reader:
        if line_count == 0:
            pass
        else:
            clean_firstname.append(row[0])
            clean_lastname.append(row[1])
            dirty_birthmonth.append(row[2])
            clean_birthday.append(row[3])
            dirty_birthyear.append(row[4])
            dirty_course.append(row[5])
            dirty_phonenumber.append(row[6])
        line_count += 1

    # cleaning birthmonth
    unique_birthmonth_dict = {'1':'January','2':'February','3':'March','4':'April','5':'May','6':'June','7':'July',
                             '8':'August','9':'September','10':'October','11':'November','12':'December','January':'January',
                              'February':'February','March':'March','April':'April','May':'May','June':'June','July':'July',
                             'August':'August','September':'September','October':'October','November':'November',
                              'December':'December','Jan':'January','Feb':'February','Mar':'March','Apr':'April',
                              'May':'May','Jun':'June','Jul':'July','Aug':'August','Sep':'September',
                              'Oct':'October','Nov':'November','Dec':'December'}
    clean_birthmonth = [re.sub(r''.join(month),r''.join(unique_birthmonth_dict[month]),month) for month in dirty_birthmonth]
    
    # cleaning birthyear
    unique_birthyear_dict = {'97':'1997','98':'1998','99':'1999','00':'2000','01':'2001','1997':'1997','1998':'1998',
                             '1999':'1999','2000':'2000','2001':'2001'}
    clean_birthyear = [re.sub(r''.join(year),r''.join(unique_birthyear_dict[year]),year) for year in dirty_birthyear]
    
    # cleaning course
    # did this first:
    # unique_course_dict = {course:None for course in set(dirty_course)}
    # for course in unique_course_dict:
        # final_course = input(course + ": " + "Enter course: ")
        # unique_course_dict[course] = final_course
    # then copied the output here:
    unique_course_dict = {'ME': 'BS Management Engineering','BS LM': 'BS Legal Management','MGT': 'BS Management',
                         'Communications Technology': 'BS Communications Technology Management','BS MGT': 'BS Management',
                         'Information Technology Entrepreneurship': 'BS Information Technology Entrepreneurship',
                         'BS CTM': 'BS Communications Technology Management','ITE': 'BS Information Technology Entrepreneurship',
                         'BS ITE': 'BS Information Technology Entrepreneurship','Management': 'BS Management',
                         'CTM': 'BS Communications Technology Management','LM': 'BS Legal Management','Legal Management': 'BS Legal Management',
                         'BS ME': 'BS Management Engineering','Management Engineering': 'BS Management Engineering'}
    clean_course = [re.sub(r''.join(course),r''.join(unique_course_dict[course]),course) for course in dirty_course]
    
    # cleaning phonenumber
    slightly_clean_phonenumber = [re.sub(r'^(9[0-9][0-9])',r'+63 \1',number) for number in dirty_phonenumber]
    clean_phonenumber = []
    for number in slightly_clean_phonenumber:
        clean_phonenumber.append(number.replace(" ",""))
        
    # combining cleaned lists
    clean_lists = list(zip(clean_firstname,clean_lastname,clean_birthmonth,clean_birthday,clean_birthyear,clean_course,clean_phonenumber))

# creating clean csv
with open('cleanrecweekdata.csv','w') as clean:
    clean_writer = csv.writer(clean,lineterminator='\n')
    clean.write('firstname,lastname,birthmonth,birthday,birthyear,course,phonenumber\n')
    clean_writer.writerows(clean_lists)
````
